```python
def make_timeseries_by_day(df):
    time_day = []
#     year = int(str(df.iloc[0,0]).split('-')[0])
    year = int(str(df.iloc[0,0])[0:4])
    print(year)
    
    if df.shape[0] == 1 :
        ori_time = str(df.iloc[0,0])+ "-06-15 12:00"
        ori_time = datetime.datetime.strptime(ori_time, '%Y-%m-%d %H:%M')
        df.iloc[0,0] = ori_time
        
        return df
    
    if df.shape[0] == 2 :
        ori_time = str(df.iloc[0,0])+ "-03-15 12:00"
        ori_time = datetime.datetime.strptime(ori_time, '%Y-%m-%d %H:%M')
        df.iloc[0,0] = ori_time
        ori_time = str(df.iloc[1,0])+ "-09-15 12:00"
        ori_time = datetime.datetime.strptime(ori_time, '%Y-%m-%d %H:%M')
        df.iloc[1,0] = ori_time
        
        return df
    
        
    div_cnt = df[df.columns[0]].value_counts().sort_index().tolist()
#     if div_cnt.shape[0] == 1 :
#         print('shape in')
#         div_cnt[0] = 1
#         div_cnt[1] = 1
            
    for i in range(len(div_cnt)):
#         print(i)
        month = calendar.monthrange(year,i+1)
        for j in range(div_cnt[i]):
    #         print(j)
    #         print(month)
            time_val = datetime.timedelta(days = (month[1]/div_cnt[i]*j)+1)
#             print('time_val' , time_val)
    #         time_day.append()
#             print(str(time.strftime('%H', time.gmtime(time_val.seconds))))
#             time_day.append('-'+str(time_val.days)+' ' + str(time.strftime('%H:%M:%S', time.gmtime(time_val.seconds))))
            time_day.append('-'+str(time_val.days)+' ' + str(time.strftime('%H:00:00', time.gmtime(time_val.seconds))))
#             print('append')
            
#     df.iloc[:,0]= ftime
#     print(time_day)
#     print('end 1')
#     print(df.iloc[:,0])
    df.iloc[:,0] = df.iloc[:,0] + time_day
#     print('end 2')
    
    return df
```

```python
def make_timeseries(df):
#     print(interpolate)
#     print(df.shape)
    df = df.dropna(axis='rows', thresh=3)
    date_col = df.columns[0]
    df[date_col] = pd.to_datetime(df[date_col])
    df = df[df[date_col].notna()]

    year = pd.DatetimeIndex(df[date_col]).year.astype(np.int64)

    start = str(year[0]) + "-01-01 00:00"
#     start

    end = str(year[0]) + "-12-31 23:00"
#     end

    time_series = pd.date_range(start = start, end = end, freq='H')
    time_series
    time_series = pd.DataFrame(time_series)
    time_series.columns = [date_col]   
    time_series

    df = pd.concat([time_series, df], axis=0)
    df = df.drop_duplicates([date_col], keep="last")
    df = df.sort_values([date_col], axis=0)
    
#     if interpolate :
# #         print('interpolation')
#         df = df.interpolate(method='spline', order=2, limit_direction='both')
        
# #     print(df.shape)
    return df
```

```python
def createDataFrame(folder, file_list, iloc_val= None, interpolate=[False,False]):
    print(folder)
# #     print(file_list)
    print(interpolate)
#     print(interpolate[0])
#     print(interpolate[1])
    day = 24 * 60 * 60
    week = day * 7
    year = (365.2425) * day

    da = db = [[] for i in range(len(file_list))]
    for i in range(len(file_list)):
        for filename in file_list[i]:
            print(filename)
            path = os.path.join(folder, filename)
            df_tmp = pd.read_excel(path)
            
            if interpolate[1]: 
                print('make_timeseries_by_day')
                df_tmp = make_timeseries_by_day(df_tmp)
            
            df_tmp = make_timeseries(df_tmp)
            
            df_iloc = eval(f"df_tmp.iloc[{iloc_val}]")
            
            df_time = pd.DataFrame(df_tmp.iloc[:,0])  
            df_full_withtime = pd.concat([df_time, df_iloc], axis = 1)
            da[i].append(df_full_withtime)
            
#             if i == 0:
#                 df_time = pd.DataFrame(df_tmp.iloc[:,0])  
#                 df_full_withtime = pd.concat([df_time, df_iloc], axis = 1)
#                 da[i].append(df_full_withtime)
#             else :
#                 da[i].append(df_iloc)

        db[i] = pd.concat(da[i])
        db[i] = db[i].reset_index(drop=True)
        
        
        if interpolate[0] :
            print('interpolation')
            for j in range(1, db[i].shape[1], 1):
#                 print('j')
#                 print(j)
                idx = db[i].iloc[:, j].dropna()
#                 print(idx)

                if idx.shape[0] != 0:
                    db[i].iloc[0, j] = idx.iloc[0]
                    db[i].iloc[-1, j] = idx.iloc[-1]
                
            db[i] = db[i].interpolate(method='polynomial', order=3, limit_direction='both')
#             db[i] = db[i].interpolate(method='pchip', order=3, limit_direction='both')

        

        date_time = pd.to_datetime(db[i].iloc[:, 0], format='%Y.%m.%d %H:%M',utc=True)
        timestamp_s = date_time.map(datetime.datetime.timestamp)
        db[i]['Day sin'] = np.sin(timestamp_s * (2 * np.pi / day))
        db[i]['Day cos'] = np.cos(timestamp_s * (2 * np.pi / day))
        # df['Week sin'] = np.sin(timestamp_s * (2 * np.pi / week))
        # df['Week cos'] = np.cos(timestamp_s * (2 * np.pi / week))
        db[i]['Year sin'] = np.sin(timestamp_s * (2 * np.pi / year))
        db[i]['Year cos'] = np.cos(timestamp_s * (2 * np.pi / year))

            
#     print(df.shape)pchip


    for i in range(1 , len(db)):
        db[i] = db[i].iloc[:,1:].rename(columns = lambda x: x + '_'+str(i))
      
    df = pd.concat(db, axis=1)

#     date_time = pd.to_datetime(df.iloc[:, 0], format='%Y.%m.%d %H:%M',utc=True)
#     timestamp_s = date_time.map(datetime.datetime.timestamp)
#     df['Day sin'] = np.sin(timestamp_s * (2 * np.pi / day))
#     df['Day cos'] = np.cos(timestamp_s * (2 * np.pi / day))
#     # df['Week sin'] = np.sin(timestamp_s * (2 * np.pi / week))
#     # df['Week cos'] = np.cos(timestamp_s * (2 * np.pi / week))
#     df['Year sin'] = np.sin(timestamp_s * (2 * np.pi / year))
#     df['Year cos'] = np.cos(timestamp_s * (2 * np.pi / year))
    
    return df
```

```python
import pandas as pd
import numpy as np
import calendar
import os
import datetime
import time
```

24회차 이상으로 회차가 있는 데이터 정리...

```python
# 녹조 조류 모니터링

def make_timeseries_by_day(df):
    time_day = []
    year = int(df.iloc[0,0].split('-')[0])
    print(year)

    
    div_cnt = df[df.columns[0]].value_counts().sort_index()
            
    for i in range(div_cnt.shape[0]):
    #     print(i)
        month = calendar.monthrange(year,i+1)
        for j in range(div_cnt[i]):
    #         print(j)
    #         print(month)
            time_val = datetime.timedelta(days = (month[1]/div_cnt[i]*j)+1)
    #         time_day.append()
#             print(str(time.strftime('%H', time.gmtime(time_val.seconds))))
#             time_day.append('-'+str(time_val.days)+' ' + str(time.strftime('%H:%M:%S', time.gmtime(time_val.seconds))))
            time_day.append('-'+str(time_val.days)+' ' + str(time.strftime('%H:00:00', time.gmtime(time_val.seconds))))

            
#     df.iloc[:,0]= ftime

    df.iloc[:,0] = df.iloc[:,0] + time_day
    
    return df
```

```python
file_name = 'young2/조류/송촌보_2017.xlsx'

df = pd.read_excel(file_name)
```

```python
df = make_timeseries_by_day(df)
```

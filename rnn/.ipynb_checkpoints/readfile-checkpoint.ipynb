```python
df
```

```python

#date_name = '일시'
date_name = 'dt'
idx = df.columns.get_loc(date_name)
columns = list(df.columns)
columns[0], columns[idx] = columns[idx], columns[0]
df=df.reindex(columns=columns)
```

```python
df
```

```python

start = df.iloc[0][0]
end = df.iloc[-1][0]

year = start.split('-')
start = year[0] + "-01-01 00:00"
end = year[0] + "-12-31 23:00"


ts_ms = pd.date_range(start = start, end = end, freq='H')
ts_ms = pd.DataFrame(ts_ms)
ts_ms.columns = [date_name]
```

```python

df[date_name] = pd.to_datetime(df[date_name])

df = pd.concat([ts_ms, df], axis=0)
df = df.drop_duplicates([date_name], keep="last")
df = df.sort_values([date_name], axis=0)
```

```python
df
```

```python
df.to_excel(output_file_name, index = False)
```
